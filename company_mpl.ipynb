```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.available
```

```python
plt.style.use("ggplot")
```

### Exercise 1


```python
data = pd.read_csv("company_sales_data.csv")
data.head()
```

```python
month_sales = data.groupby(["month_number"])["total_profit"].sum()
month_sales
```

```python
plt.plot(month_sales.index, month_sales.values, color="k")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.title("Total Profit of All Months")
```

### Exercise 2


```python
month_units = data.groupby(["month_number"])["total_units"].sum()
month_units
```

```python
plt.plot(
    month_units.index,
    month_units.values,
    marker="o",
    linestyle="dotted",
    color="r",
    linewidth=3,
    label="Total Profit Over Time",
)
plt.legend(loc="lower right")
plt.xlabel("Month Number")
plt.ylabel("Sold units number")
plt.title("Total Profit of All Months")
```

### Exercise 3


```python
# facecream 	facewash 	toothpaste 	bathingsoap 	shampoo 	moisturizer
```

```python
plt.plot(data.month_number, data.facecream, label="facecream", linestyle="-")
plt.plot(data.month_number, data.facewash, label="facewash", marker="o", linestyle=":")
plt.plot(data.month_number, data.toothpaste, label="toothpaste", linestyle="--")
plt.plot(data.month_number, data.bathingsoap, label="bathingsoap", linestyle="-.")
plt.plot(data.month_number, data.shampoo, label="shampoo", linestyle="--")
plt.plot(data.month_number, data.moisturizer, label="moisturizer")
plt.legend(loc="upper left")
plt.xlabel("Month Number")
plt.ylabel("Sold units number")
plt.title("All Product Sales")
```

### Exercise 4


```python
toothpaste_sales = data.groupby("month_number")["toothpaste"].sum()
toothpaste_sales
```

```python
plt.scatter(toothpaste_sales.index, toothpaste_sales.values)
plt.title("Monthly Toothpaste Sales")
plt.xlabel("Month Number")
plt.ylabel("Toothpaste Sold units number")
```

### Exercise 5


```python
ax = plt.axes(label="5")
width = 0.4
plt.bar(data.month_number, data.facewash,
        color="r", label="Facewash", width=0.25)
plt.bar(
    data.month_number + width,
    data.facecream,
    color="forestgreen",
    label="Facecream",
    width=0.25,
)
plt.legend(loc="upper left")
ax.set(
    xlabel="Month Number",
    ylabel="Number of Units Sold",
    title="Facewash and Facecream Total Units Sold",
)
```
